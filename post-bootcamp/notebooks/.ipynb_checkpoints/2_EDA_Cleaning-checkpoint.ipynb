<center><h1>Exploratory Data Analysis</h1></center>

The purpose of this notebook will be to discover what factors most strongly correlate to a player's fantasy performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

# hs for history seasons
hs_df = pd.read_csv('../data/final_history.csv', index_col=0)
```

We will begin with the aggregated historical data from seasons 2006/07 - 2022/23. Some players may not have data for the entire range due to transfers (i.e. not being in the league at the time therefore having no data).

We should begin by cleaning the data and reducing repetitive or unnecessary columns. 
In the history seasons dataframe, only the `element_code` column is now repetitive and unnecessary in our current objective since we have already matched this code to the player's name.

```python
hs_df = hs_df.drop(['element_code'], axis=1)
print(f"{hs_df.shape}")
```

```python
hs_df.head()
```

Let's see the relationship between `total_points` earned in a fantasy season and the overall `ict_index`  which combines statistics for a player's influence, creativity and threat. As well as the individual `influence`, `creativity`, and `threat` columns.

```python
def scatt(col, dataframe):
    '''This function helps to create a scatterplot for any variable against total_points.'''
    assert isinstance(col, str), "Input should be string"
    assert isinstance(dataframe, pd.DataFrame)
    assert col in dataframe.columns, "Input should exist in dataframe"
    assert 'total_points' in dataframe.columns, "total_points must exist in dataframe"
    plt.figure()
    plt.scatter(dataframe[f"{col}"], dataframe['total_points'])
    plt.xlabel(f"{col}".title())
    plt.ylabel('Total Points')
    plt.title(f"Total Points vs {col}".title())
    plt.show()
    
ict_stats = ['influence', 'creativity', 'threat', 'ict_index']

for col in ict_stats:
    scatt(col, hs_df)
```

There are some slight differences `influence` compared to `creativity` and `threat` where `influence` has a more linear relationship with `total_points` and the distributions for `creativity` and `threat` are more similar. The overall `ict_index` seems to favour the distribution of `influence` this may be a reason to include all three statistics separately in the machine learning part of this project (this will be further explored later with a correlation map).

The general trend is that the higher the `ict_index` the greater amount of points earned by the player. Although there are signficant outliers when the ICT index is 0, these seem to be wildcards where there may not have been enough prior information on the player. Let's check out these columns in our dataframe.

```python
hs_df.loc[hs_df['ict_index'] == 0, ['name','position','season_name', 'start_cost', 'end_cost', 'total_points', 'influence', 'creativity', 'threat', 'ict_index']].sort_values('total_points', ascending=False)
```

From the above discovery we can see that these values are incorrectly inputted into the FPL API, the players who score high do not have any `ict_index` scores. Additionally, majority of the missing values are likely from newly promoted teams or new transfers to the Premier League where their is no available data (for FPL).

We can impute these values based on the average `influence`, `creativity`, `threat`, `ict_index` of players with a similar `total_points` output and `end_cost`.

```python
to_impute = hs_df.loc[hs_df['ict_index'] == 0,]
```

```python
to_impute
```

It seems that values for influence, creativity, threat, ict_index, starts, expected_goals, expected_assists, expected_goal_involvements, and expected_goals_conceded are missing for all of these players.

We will attempt to impute the missing values based on other players with similar cost and total_point outputs.

```python
col_to_impute = hs_df.columns[-9:]
col_to_impute
```

```python
# for the missing values we will take the cost and total_points of that player for the season and check for similar rows
# in the original dataframe based on a range of the average difference in cost and average difference in points earned.
# if these values exist we can impute the average of the similar rows ICT index to the missing player
for index, row in to_impute.iterrows():
    cost = row['end_cost']
    total_points = row['total_points']
    similar_rows = hs_df[(hs_df['end_cost'] >= cost - 1) & \
                         (hs_df['end_cost'] <= cost + 1) & \
                         (hs_df['total_points'] >= total_points - 30) & \
                         (hs_df['total_points'] <= total_points + 30) & \
                         (hs_df['ict_index'] != 0)]
    if len(similar_rows) > 0:
        # Imputing missing influence, creativity, threat, ict_index, starts,
        # expected_goals, expected_assists, expected_goal_involvements, expected_goals_conceded
        for col in col_to_impute:
            hs_df.loc[index, f'{col}'] = similar_rows[f'{col}'].mean()
```

Let's check if there are any rows that do not have any matches and therefore must be dropped from the dataframe.

```python
hs_df.loc[hs_df['ict_index']==0,]
```

There are 2 rows that could not find similar players to impute to either due to higher cost of player or having a abnormally high total_points.

```python
hs_df = hs_df.loc[hs_df['ict_index'] != 0].reset_index(drop = True)

print(f"{hs_df.shape}")
```

As we can see from the above we have now correctly imputed all of the values based on players of similar end cost and total_points achieved. Let's check what the `ict_index` vs `total_points` graph distributions appear as now. 

```python
for col in ict_stats:
    scatt(col, hs_df)
```

The scoring system in FPL changes based on position. The <a href="https://www.premierleague.com/news/2174909">article</a> posted by FPL gives a full description of the scoring system.

Let's see how `ict_stats` are distributed on average by position.

```python
def avgbypos(stat, dataframe):
    '''This function helps to find average statistics by position and plots them in a bar graph'''
    assert isinstance(stat, str), "Input should be a string"
    assert stat in dataframe.columns, "Input string does not exist in the DataFrame"
    # Finds statistic averages by position
    pos_avg = []
    for x in dataframe['position'].unique():
        pos_avg.append(dataframe.loc[hs_df['position'] == x, f"{stat}"].mean())

    # plots the averages by position
    plt.figure()
    plt.bar(dataframe['position'].unique(), pos_avg)
    plt.title(f"average {stat} by position".title())
    plt.xlabel("Position")
    plt.ylabel(f"{stat}".title())
    plt.show()
```

```python
for col in ict_stats:
    avgbypos(col, hs_df)
```

The forwards and midfielders have the greatest `creativity` and `threat` values since they are more involved in offensive contributions such as goals, assists and key passes. Additionally, the FPL scoring system as linked above gives priority to attacking contributions compared to defensive. 

`Influence` averages seem to be similar across all positions with slight favour to defenders and forwards. 

The overall `ict_index` values favour midfielders and forwards because of their attacking contributions and `influence` statistics being similar across all players.

This was also discovered in the previous notebooks where forwards and midfielders are the most valuable positions to invest in.

Following this discovery, let's attempt to determine the average total points, minutes, goals scored, assists, clean sheets, and goals conceded by position.

```python
game_stats = ['minutes', 'goals_scored', 'assists', 'clean_sheets', 'goals_conceded']

for col in game_stats+['total_points']:
    avgbypos(col, hs_df)
```

It seems that the greatest `total_points`, `goals_scored` and `assists` tallies come from forwards followed by midfielders, whereas the greatest `minutes`, `clean_sheets`, and `goals_conceded` are similar across all positions with slight favour to defenders. 

This helps us further confirm that forwards and midfielders are the best to invest in to win FPL.
